```python
print("Programa que dado un texto proporcionado por el usuario, imprime el número de veces que cada palabra se repite.\n\n")


texto = input("Introduce una frase:  ")                         # texto introducido por el usuario con input
print("\n\n  La frase que introduciste es: \n\n", texto)        


mi_lista = texto.split()                                        #convierte una cadena a lista
print("\n\n Tu frase convertida en lista es:   " ,mi_lista)

print("\n\n La palabra 'El' se repite: " , mi_lista.count("El"), "instancias")
print("\n\n La palabra 'amor' se repite: " , mi_lista.count("amor"), "instancias")

# Aquí podríamos imprimir una a una las líneas de la frase pero no es el caso.
# Mas adelante otra manera de hacerlo creando un diccionario

print("\n\n ----------------------------------------------------------------------------------")
print("\n\n ----------------------------------------------------------------------------------")
print("\n\n ----------------------------------------------------------------------------------")

frecuencias = {}                          # declaramos un diccionario para ir almacenando las palabras

for palabra in mi_lista:                  #recorremos con for cada palabra en la lista
    if palabra in frecuencias:
        frecuencias[palabra] += 1         # la palabra ya está y se le suma 1
    else:
        frecuencias[palabra] = 1          # la palabra no está, se le suma 1
        
        
        
for palabra in frecuencias:                # recorremos el diccionario
    cantidad = frecuencias[palabra]
    print(f"\n\n La palabra '{palabra}' se repite {cantidad} instancia(s)")     # imprimimos
    
    
    
```
