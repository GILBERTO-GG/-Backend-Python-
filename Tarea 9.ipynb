```python
import pandas as pd

df = pd.read_csv('titanic.csv')
print(df)
```

```python

#--->a) El Histograma de frecuencias para la variable edad

import matplotlib.pyplot as plt
hg, ax = plt.subplots()
ax.hist(df["Age"]) 
ax.set_xlabel("EDADES")
ax.set_ylabel("TOTALES")
plt.show()
```

```python
hg.savefig('histograma.jpg', dpi=300)
```

```python

#--->b) El diagrama de caja para la variable edad

import matplotlib.pyplot as plt
fig, ax = plt.subplots()
ax.boxplot([df["Age"]])   
ax.set_xticklabels(["Edades"])
ax.set_xlabel("# Totales")
plt.show()
```

```python
fig.savefig('diagrama_caja.jpg', dpi=300)
```

```python

#--->c) Un grfico que muestre el nmero de pasajeros de cada clase (variable Pclass)

import pandas as pd
import matplotlib.pyplot as plt




data = pd.read_csv('titanic.csv')


passenger_count_by_class = data['Pclass'].value_counts().sort_index()

# Crear el grfico de barras
plt.bar(passenger_count_by_class.index, passenger_count_by_class.values)
plt.xlabel('Clase')
plt.ylabel('Nmero de Pasajeros')
plt.title('Nmero de Pasajeros por Clase en el Titanic')
plt.xticks(passenger_count_by_class.index)
plt.show()
```

```python
plt.savefig('grafico_titanic.jpg', format='jpg', dpi=700)
```
