```python
import pandas as pd

df = pd.read_csv('Studentsperformance.csv')
df
```

# Creación de un stripplot (Gráfico de Franjas / Bandas)

```python
import seaborn as sns
sns.stripplot(data = df, y = "writing score", x = "math score", jitter = True) #--> jitter = colores a las intensidades
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()  #--> sentencia de matplotlib
sns.stripplot(data = df, y = "writing score", x = "math score", jitter = True)
ax.axes.xaxis.set_ticks([])  #---> indica que no queremos comentarios numéricos dentro del eje X
plt.show
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
sns.stripplot(data = df, y = "writing score", x = "race/ethnicity", jitter = True)
plt.show
```

# Creación de un swarmplot (Diagrama de enjambre)
Busca colocar las observaciones de manera que no se empalmen

```python
sns.swarmplot(data = df, y = "writing score", x = "race/ethnicity")
```

# Creación de un Boxplot (Diagrama de caja)

```python
sns.boxplot(data = df, y = "writing score", x = "race/ethnicity")
```

# Creación de gráfico de violín 
Mezcla de un boxplot con función de densidad Kernel

```python
sns.violinplot(data = df, y = "writing score", x = "race/ethnicity")
```

# Creación de un lvplot (Letter-Value Plot)
Mezcla del Boxplot y el Violinplot, conveniente para bases de datos grandes

```python
sns.catplot(x = "race/ethnicity", y = "writing score", kind = "boxen", data = df)
```

```python
import seaborn as sns
import pandas as pd

df = pd.read_csv('Studentsperformance.csv')
df
```

# Creación de un mapa de calor para una matriz de correlación 

```python
sns.heatmap(df.corr())
```

```python
sns.heatmap(df.corr(), annot = True)
```

# Creación de una tabla cruzada

```python
pd.crosstab(df["gender"], df["race/ethnicity"], values = df["math score"], aggfunc = "mean")
```

# Creación de un mapa de calor para una tabla cruzada

```python
sns.heatmap(pd.crosstab(df["gender"], df["race/ethnicity"], values = df["math score"], aggfunc = "mean"), annot = True)
```

# Creación de gráficos compuestos

# Uso de facetGrid

```python
import seaborn as sns
graf = sns.FacetGrid(df, col = "race/ethnicity")
graf.map(sns.boxplot, "math score", order = ["group A", "group B", "group C", "group D", "group E"])
```

# Uso de FactorPlot

```python
sns.factorplot(x = "math score", data = df, col = "race/ethnicity", kind = "box")
```

# FacetGrid para gráfico de dispersión

```python
import matplotlib.pyplot as plt
graf = sns.FacetGrid(df, col = "race/ethnicity")
graf.map(plt.scatter, "reading score", "writing score")
```

# Gráficos tabulares de regresión

```python
sns.lmplot(data = df, x = "reading score", y = "writing score", col = "gender", row = "race/ethnicity")
```

# Creación de una gráfica conjunta

```python
graf = sns.JointGrid(data = df,  x = "reading score", y = "writing score")
graf.plot(sns.regplot, sns.distplot)
```

```python
sns.jointplot(data = df, x = "reading score", y = "writing score", kind = "hex")
```

```python
graf = sns.jointplot(data = df.query('gender == "male"'), x = "reading score", y = "writing score", kind = "scatter")
```
