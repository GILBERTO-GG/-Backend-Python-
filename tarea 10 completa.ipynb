```python
import pandas as pd        #importe de los datos del archivo en un data frame

df = pd.read_csv('SYB64_246_202110_Population Growth, Fertility and Mortality Indicators.csv', encoding='latin1',skiprows=1)
df
```

```python
df["Series"]       #todos los conceptos de la serie
```

```python
df["Series"][2]    # Mortalidad infantil para ambos sexos
```

```python
df["Series"][4]   #Esperanza de vida al nacer para ambos sexos
```

```python
# tasa de mortalidad infantil por el aÃ±o 2020
df2 = df[df["Series"] == "Infant mortality for both sexes (per 1,000 live births)"]
df2 = df2[df2["Year"] == 2020]
df2 = df2[["Unnamed: 1", "Value"]]
df2.rename({'Unnamed: 1': 'Country', 'Value' : 'Infant mortality'}, axis=1, inplace= True)
df2 = df2.set_index("Country")
df2
```

```python
#esperanza de vida al nacer por el aÃ±o 2020
df3 = df[df["Series"] == "Life expectancy at birth for both sexes (years)"]
df3 = df3[df3["Year"] == 2020]
df3 = df3[["Unnamed: 1", "Value"]]
df3.rename({"Unnamed: 1": "Country", "Value" : "Life expectancy"}, axis=1, inplace= True)
df3 = df3.set_index("Country")
df3
```

```python
# Tasa de aumento anual de la poblaciÃ³n por el aÃ±o 2020
df4 = df[df["Series"] == 'Population annual rate of increase (percent)']
df4 = df4[df4["Year"] == 2020]
df4 = df4[["Unnamed: 1", "Value"]]
df4.rename({'Unnamed: 1': 'Country', 'Value' : 'Population increase'}, axis =1, inplace = True)
df4 = df4.set_index('Country')
df4
```

```python
# Tasa total de fecundidad por el aÃ±o 2020
df5 = df[df["Series"] == 'Total fertility rate (children per women)']
df5 = df5[df5["Year"] == 2020]
df5 = df5[["Unnamed: 1", "Value"]]
df5.rename({'Unnamed: 1': 'Country', 'Value' : 'Fertility rate'}, axis =1, inplace = True)
df5 = df5.set_index('Country')
df5
```

# No hay coincidencias para el aÃ±o 2020

```python
#Tasa de mortalidad materna por el aÃ±o 2020
df6 = df[df["Series"] == 'Maternal mortality ratio (deaths per 100,000 population)']
df6 = df6[df6["Year"] == 2020]
df6 = df6[["Unnamed: 1", "Value"]]
df6.rename({'Unnamed: 1': 'Country', 'Value' : 'Maternal mortality'}, axis =1, inplace = True)
df6 = df6.set_index('Country')
df6
```

```python
# Esperanza de vida al nacer para hombres 
df7 = df[df["Series"] == 'Life expectancy at birth for males (years)']
df7 = df7[df7["Year"] == 2020]
df7 = df7[["Unnamed: 1", "Value"]]
df7.rename({'Unnamed: 1': 'Country', 'Value' : 'Life expectancy at birth for males'}, axis =1, inplace = True)
df7 = df7.set_index('Country')
df7
```

```python
# Esperanza de vida al nacer para mujeres 
df8 = df[df["Series"] == 'Life expectancy at birth for females (years)']
df8 = df8[df8["Year"] == 2020]
df8 = df8[["Unnamed: 1", "Value"]]
df8.rename({'Unnamed: 1': 'Country', 'Value' : 'Life expectancy at birth for females'}, axis =1, inplace = True)
df8 = df8.set_index('Country')
df8
```

```python
grf = pd.concat([df2, df3, df4, df5, df7, df8], join = "outer", axis = 1)
grf = grf[1:]
grf = grf.dropna()
grf
```

```python
grf["Infant mortality"] = pd.to_numeric(grf["Infant mortality"], downcast = "float")
grf["Life expectancy"] = pd.to_numeric(grf["Life expectancy"], downcast = "float")
grf["Population increase"] = pd.to_numeric(grf["Population increase"], downcast = "float")
grf["Fertility rate"] = pd.to_numeric(grf["Fertility rate"], downcast = "float")
grf["Life expectancy at birth for males"] = pd.to_numeric(grf["Life expectancy at birth for males"], downcast = "float")
grf["Life expectancy at birth for females"] = pd.to_numeric(grf["Life expectancy at birth for females"], downcast = "float")
```

# RelaciÃ³n entre tasa de mortalidad infantil y esperanza de vida al nacer por el aÃ±o 2020

```python
import seaborn as sns 
sns.regplot(data = grf, y = "Infant mortality", x = "Life expectancy", marker = "*")
```
