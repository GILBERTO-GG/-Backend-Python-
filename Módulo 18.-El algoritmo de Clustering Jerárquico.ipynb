```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os 
```

```python
os.chdir("/Users/gilgu/OneDrive/Escritorio")
data = pd.read_csv('Wholesale customers data.csv')
data.head()
```

```python
from sklearn.preprocessing import normalize 
data_scaled = normalize(data)
data_scaled
```

```python
data_scaled = pd.DataFrame(data_scaled, columns = data.columns)      #---> convierte un arreglo en un DataFrame
data_scaled.head()
```

```python
import scipy.cluster.hierarchy as shc
plt.figure(figsize=(10,7))                                               #---> definir un tamaño de figura para el gráfico
plt.title("Dendograma")                                                  #--->Definir un título
dend = shc.dendrogram(shc.linkage(data_scaled, method= 'ward'))          #--->escala de Ward(agrupación para grupos) 
```

```python
print(dend['color_list']) #>imprime la lista de colores asociados a cada uno de los puntos(observaciones) en el dendrograma
```

```python
colores_unicos = set(dend['color_list'])
colores_unicos
```

```python
num_clusters_optimo = len(colores_unicos) -1
num_clusters_optimo 
```

```python
plt.figure(figsize=(10,7))
plt.title("Dendograma")
dend = shc.dendrogram(shc.linkage(data_scaled, method = 'ward'))
plt.axhline(y=6, color = 'r', linestyle = '--')
```

```python
data
```

```python
from sklearn.cluster import AgglomerativeClustering
cluster = AgglomerativeClustering(n_clusters=2, affinity='euclidean', linkage='ward')
grupos = cluster.fit_predict(data_scaled)
grupos
```

```python
plt.figure(figsize=(10,7))
plt.scatter(data_scaled['Milk'], data_scaled['Grocery'], c = cluster.labels_)
```

# Análisis del caso de Estudiantes.

```python
data = pd.DataFrame(np.array([[7, 6.5, 9.2, 8.6, 8],
                            [7.5, 9.4, 7.3, 7, 7],
                            [7.6, 9.2, 8, 8, 7.5],
                            [5, 6.5, 6.5, 7, 9],
                            [6, 6 , 7.8, 8.9, 7.3],
                            [7.8, 9.6, 7.7, 8, 6.5],
                            [6.3, 6.4, 8.2, 9, 7.2],
                            [7.9, 9.7, 7.5, 8, 6],
                            [6, 6, 6.5, 5.5, 8.7],
                            [6.8, 7.2, 8.7, 9, 7]]),
                  columns = ['Matematicas','Ciencias','Español','Historia','EdFisica'])

data
```

```python
from sklearn.preprocessing import normalize
data_scaled = normalize(data)
data_scaled
```

```python
import scipy.cluster.hierarchy as shc
plt.figure(figsize=(10,7))
plt.title("Dendograma")
dend = shc.dendrogram(shc.linkage(data_scaled, method = 'ward'))
```

```python
colores_unicos = set(dend['color_list'])
num_clusters_optimo = len(colores_unicos) -1
num_clusters_optimo 
```

```python
plt.figure(figsize=(10,7))
plt.title("Dendograma")
dend = shc.dendrogram(shc.linkage(data_scaled, method = 'ward'))
plt.axhline(y=0.2, color = 'r', linestyle = '--')
```

# Análisis gráfico con PCA

```python
campos = data_scaled
from sklearn import decomposition
pca = decomposition.PCA(n_components = 2)
pca.fit(campos)
campos = pca.transform(campos)
campos
```
