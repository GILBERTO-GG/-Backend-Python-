```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt 
import os 

os.chdir("/Users/gilgu/OneDrive/Escritorio")
```

```python
data = pd.read_csv("Breast-cancer.csv")
data
```

```python
data.drop(["id","Unnamed: 32"], axis = 1, inplace = True)
data.head()
```

```python
M = data[data.diagnosis == "M"]
B = data[data.diagnosis == "B"]
```

# Diagrama de dispersión 

```python
plt.scatter(M.radius_mean, M.texture_mean, color = "red", label = "Maligno", alpha = 0.3)
plt.scatter(B.radius_mean, B.texture_mean, color = "green", label = "Benigno", alpha = 0.3)
plt.xlabel("radius_mean")
plt.ylabel("texture_mean")
plt.legend()
plt.show
```

#  0 = Benigno,   1 = Maligno

```python
data.diagnosis = [1 if each == 'M' else 0 for each in data.diagnosis]
data
```

```python
y = data.diagnosis.values
x_data = data.drop(["diagnosis"], axis = 1)
```

```python
y
```

```python
x_data
```

# Normalización:

```python
x = (x_data - np.min(x_data)) / (np.max(x_data) - np.min(x_data))
x
```

# Bases de entrenamiento y prueba:

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.30, random_state=1)
```

# Modelo KNN (K vecinos más cercanos):

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(x_train, y_train)
prediction = knn.predict(x_test)
print("Score: ", knn.score(x_test, y_test))
```

```python
y_pred = prediction
y_true = y_test
```

# Creación de la matriz de confusión:

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_true, y_pred)
cm
```

# Visualización de la matriz de confusión

```python
import seaborn as sns
f, ax = plt.subplots(figsize = (5,5))

sns.heatmap(cm, annot = True, linewidths = 0.5, linecolor = "red", fmt = ".0f", ax = ax)
plt.xlabel("y pronosticada")
plt.ylabel("y verdadera")
plt.show()
```

# Cálculo de la precisión global:

```python
Correctos = cm[0,0] + cm[1,1]
Incorrectos = cm[0,1] + cm [1,0]
PrecisionGlobal = Correctos / (Correctos + Incorrectos)
PrecisionGlobal
```

```python
PrecisionBenigno = cm[0,0] / (cm[0,0] + cm[1,0])
PrecisionBenigno
```

```python
PrecisionMaligno = cm[1,1] / (cm[1,1] + cm[0,1])
PrecisionMaligno
```

```python
from sklearn.metrics import roc_curve
from sklearn.metrics import auc

y_scores = knn.predict_proba(x_test)

fpr, tpr, threshold = roc_curve(y_test, y_scores[:,1])
# fpr = False positive rate,  tpr = True positive rate
roc_auc = auc(fpr, tpr)

plt.title('Receiver Operating Characteristic')
plt.plot(fpr, tpr, 'b', label = 'AUC = %0.2f' % roc_auc)
plt.legend(loc = "lower right")
plt.plot([0,1],[0,1],'r--')
plt.xlim([0,1])
plt.ylim([0,1])
plt.ylabel("Tasa de positivos verdaderos")
plt.xlabel("Tasa de falsos positivos")
plt.title("Curva ROC de KNN")
plt.show()
```
