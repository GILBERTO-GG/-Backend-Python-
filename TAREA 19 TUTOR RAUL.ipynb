```python
my_basket=[['bread', 'butter','wine','bananas','coffee','carrots'],
           ['tomatoes', 'onions', 'cheese', 'milk', 'potatoes'],
           ['beer', 'chips ', 'asparagus', 'salsa', 'milk', 'apples'],
           ['olive oil', 'bread', 'butter', 'tomatoes', 'steak', 'carrots'],
           ['tomatoes', 'onions', 'chips', 'wine', 'ketchup', 'orange juice'],
           ['bread', 'butter', 'beer', 'chips', 'milk'],
           ['butter', 'tomatoes', 'carrots', 'coffee', 'sugar'],
           ['tomatoes', 'onions', 'cheese', 'milk', 'potatoes'],
           ['bread', 'butter', 'ketchup', 'coffee', 'chicken wings'],
           ['butter', 'beer', 'chips', 'asparagus', 'apples'],
           ['tomatoes', 'onion', 'beer', 'chips', 'milk', 'coffee']]
```

```python
elementos = len(my_basket)
elementos
```

```python
column_names = ["Ticket", "Producto", "Cantidad"]
datos = pd.DataFrame(columns = column_names)
```

```python
contador = 0
contador2 = 0
for elemento in my_basket:
    #print(contador)
    for elemento2 in elemento:
        #print(elemento2)
        datos.at[contador2,"Ticket"] = contador
        datos.at[contador2,"Producto"] = elemento2
        datos.at[contador2,"Cantidad"] = 1
        contador2 += 1
    contador += 1
datos 
```

```python
basket = (datos.groupby(['Ticket', 'Producto']))['Cantidad'].sum().unstack().reset_index().fillna(0).set_index('Ticket')
basket
```

# CodificaciÃ³n a binario de la base previa para identificar si un artÃ­culo se comprÃ³ o nÃ³.

```python
def codificar(x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1
    
basket_codificado = basket.applymap(codificar)
basket_codificado
```

# Filtrado de aquellas transacciones en donde se ha comprado mÃ¡s de un artÃ­culo.

```python
basket_filtrado = basket_codificado[(basket_codificado > 0).sum(axis=1) >= 2]
basket_filtrado
```

# AplicaciÃ³n del algoritmo "Apriori" para localizar los artÃ­culos frecuentemente comprados en la base de datos.

```python
from mlxtend.frequent_patterns import apriori
```

# Se utiliza un "Support" mÃ­nimo del 3%

```python
articulos_frecuentes = apriori(basket_filtrado, min_support= 0.03,
                              use_colnames = True).sort_values('support', ascending = False).reset_index(drop= True)
articulos_frecuentes
```

```python
articulos_frecuentes['length'] = articulos_frecuentes['itemsets'].apply(lambda x: len(x))
articulos_frecuentes
```

```python
from mlxtend.frequent_patterns import association_rules
```

```python
association_rules(articulos_frecuentes, metric = 'lift',
                 min_threshold =1).sort_values('lift', ascending = False).reset_index(drop=True)
```
