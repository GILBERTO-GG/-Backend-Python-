```python
import pandas as pd

df = pd.read_csv('olympic-medals.csv', index_col=0)
df
```

```python
medallas = df.head(10)   #------>equipos que ganaron mas medallas.
medallas
```

# Visualización de medallas de oro y plata

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
ax.bar(medallas.index, medallas["Gold"])    #------> bar =  Gráfico de barras
plt.show()
```

# Rotar títulos a 90 grados

```python
fig, ax = plt.subplots()
ax.bar(medallas.index, medallas["Gold"])
ax.set_xticklabels(medallas.index, rotation=90) #----> esta instrucción rota a 90 grados.
ax.set_ylabel("Número de medallas")
plt.show()
```

```python
fig, ax = plt.subplots()
ax.bar(medallas.index, medallas["Gold"])
ax.bar(medallas.index, medallas["Silver"], bottom = medallas["Gold"])
ax.set_xticklabels(medallas.index, rotation=90)
ax.set_ylabel("Número de medallas")
plt.show()
```

# Visualización de medallas de Oro, Plata y Bronce.

```python
fig, ax = plt.subplots()
ax.bar(medallas.index, medallas["Gold"])
ax.bar(medallas.index, medallas["Silver"], bottom = medallas["Gold"])
ax.bar(medallas.index, medallas["Bronze"], bottom = medallas["Gold"] + medallas["Silver"] )
ax.set_xticklabels(medallas.index, rotation=90)
ax.set_ylabel("Número de medallas")
plt.show()
```

```python
fig, ax = plt.subplots()
ax.bar(medallas.index, medallas["Gold"])
ax.bar(medallas.index, medallas["Silver"], bottom = medallas["Gold"])
ax.bar(medallas.index, medallas["Bronze"], bottom = medallas["Gold"] + medallas["Silver"] )
ax.set_xticklabels(medallas.index, rotation=90)
ax.set_ylabel("Número de medallas")
ax.legend(["Gold", "Silver", "Bronze"])   #-------> aquí se colocan las leyendas
plt.show()
```

# Creación de un Histograma

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
ax.hist(df["Gold"])     #------>histograma
plt.show()
```

# Creación de dos histogramas en un solo gráfico

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
ax.hist(df["Gold"])
ax.hist(df["Silver"])
ax.set_xlabel("# de medallas")
ax.set_ylabel("# de observaciones")
plt.show()
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
ax.hist(df["Gold"], label="Oro")     #-------> aquí label pone la leyenda, va en cada renglón
ax.hist(df["Silver"], label="Plata")
ax.set_xlabel("# de medallas")
ax.set_ylabel("# de observaciones")
ax.legend()
plt.show()
```

# Creación de histogramas con especificación de intervalos

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
ax.hist(df["Gold"], label="Oro", bins = [0,5,10,15,20,25,30,35,40])    #-----> bins = intervalos
ax.hist(df["Silver"], label="Plata", bins = [0,5,10,15,20,25,30,35,40])
ax.set_xlabel("# de medallas")
ax.set_ylabel("# de observaciones")
ax.legend()
plt.show()
```

# Creación de histogramas transparentes

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
ax.hist(df["Gold"], label="Oro", bins = [0,5,10,15,20,25,30,35,40], histtype = "step")   #----> step = tipo de histograma
ax.hist(df["Silver"], label="Plata", bins = [0,5,10,15,20,25,30,35,40], histtype = "step")
ax.set_xlabel("# de medallas")
ax.set_ylabel("# de observaciones")
ax.legend()
plt.show()
```

# Creación de diagramas de caja

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
ax.boxplot([df["Gold"], df["Silver"]])   #----> boxplot = diagrama de caja
ax.set_xticklabels(["Oro","Plata"])
ax.set_xlabel("# de medallas")
plt.show()
```

# Creación de gráficos de dispersión

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
ax.scatter(df["Gold"], df["Silver"])        # scatter
ax.set_xlabel("# de medallas de oro")
ax.set_ylabel("# de medallas de plata")
plt.show()
```

# Creación de gráficos de dispersión múltiples.

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
ax.scatter(df["Gold"], df["Silver"], color = "red", label = "Plata")
ax.scatter(df["Gold"], df["Bronze"], color = "blue", label = "Bronce")
ax.set_xlabel("# de medallas de oro")
ax.set_ylabel("# de medallas de plata")
ax.legend()
plt.show()
```
