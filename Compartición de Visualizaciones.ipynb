```python
import pandas as pd

df = pd.read_csv('co2-mm-mlo_csv.csv')
print(df)
```

```python
import matplotlib.pyplot as plt
fechas = df[df["Date"] >= "2015-01-01"]

fig, ax = plt.subplots()
ax.plot(fechas["Date"], fechas["Interpolated"])
ax.axes.xaxis.set_ticks([])
plt.xlabel("Datos mensuales:  2015-01-01 - 2018-09-01")
plt.ylabel("Emisiones de CO2 promedio")
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.style.use("ggplot")   #-----------------------> le pone un color de fondo
fechas = df[df["Date"] >= "2015-01-01"]

fig, ax = plt.subplots()
ax.plot(fechas["Date"], fechas["Interpolated"])
ax.axes.xaxis.set_ticks([])
plt.xlabel("Datos mensuales:  2015-01-01 - 2018-09-01")
plt.ylabel("Emisiones de CO2 promedio")
plt.show()
```

# Re-inicialización del estilo del gráfico por "default"
