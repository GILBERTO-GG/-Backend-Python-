# Remoción de ejes

```python
sns.set_style("ticks")
sns.displot(df["reading score"], kde = True)
sns.despine(left=True)    #-----> despine= elimina los ejes en los gráficos
```

```python
sns.set_style("ticks")
sns.displot(df["reading score"], kde = True)
sns.despine(left=True, bottom= True) 
```

# Combinación de parámetros con Matplotlib

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
sns.distplot(df["writing score"], ax = ax)
ax.set(xlabel = "Score de escritura", ylabel = "Frecuencia relativa")
```

# Truncamiento de escala de visualización

```python
fig, ax = plt.subplots()
sns.distplot(df["writing score"], ax = ax)
ax.set(xlabel = "Score de escritura", ylabel = "Frecuencia relativa", xlim = (0,90), #--->límites de la escala de X
                 title = "Histograma de Frecuencias para puntajes de escritura")    #---> poner título
```

# Poner un gráfico junto a otro.

```python
import matplotlib.pyplot as plt

fig, (ax0, ax1) = plt.subplots(nrows = 1, ncols = 2, sharey = True)
sns.distplot(df["writing score"], ax = ax0)          #--------------->histograma para TODAS las observaciones
sns.distplot(df.query('gender == "male"')["writing score"], ax = ax1)  #-->histograma para los hombres
ax0.set(xlabel = "Puntaje general de escritura")
ax1.set(xlabel = "Puntaje de escritura para hombres")
ax1.axvline(x = 75, label = "Objetivo", linestyle = "--")  #--> la línea en la gráfica
ax1.legend()
```

# Asignación de colores al histograma

```python
paleta = ["deep", "muted", "pastel", "bright", "dark", "colorblind"]

for p in paleta: 
    sns.set_palette(p)
    sns.displot(df["reading score"], kde = True)
```
