```python
import os
import pandas as pd
os.chdir("/Users/gilgu/OneDrive/Escritorio")

customer_data_file = 'Online Retail.xlsx'
datos = pd.read_excel(customer_data_file)
datos.head(10)
```

# Eliminar todos los valores nulos.

```python
datos = datos.dropna()
datos.info()
```

# Delimitación de valores positivos de 'Quantity' (pues deseamos analizar lo realmente comprado).

```python
datos_reales = datos[datos['Quantity'] > 0]
datos_reales.info()
```

# consolidar por número de ticket

```python
basket = (datos_reales.groupby(['InvoiceNo', 'Description']))['Quantity'].sum().unstack().reset_index().fillna(0).set_index('InvoiceNo')
basket
```

# Codificación a binario de la base previa para identificar si un artículo se compró o no.

```python
def codificar(x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1
    
basket_codificado = basket.applymap(codificar)
basket_codificado
```

# Filtrado de aquellas transacciones en donde se ha comprado más de un artículo.

```python
basket_filtrado = basket_codificado[(basket_codificado > 0).sum(axis=1) >= 2]
basket_filtrado
```

# Aplicación del Algoritmo "a priori" para localizar los artículos frecuentemente comprados en la base de datos.

```python
from mlxtend.frequent_patterns import apriori
```

# Análisis "a priori" con un soporte mínimo del 3%

```python
articulos_frecuentes = apriori(basket_filtrado, min_support = 0.03, 
                               use_colnames = True).sort_values('support', ascending = False).reset_index(drop=True)
articulos_frecuentes['length'] = articulos_frecuentes['itemsets'].apply(lambda x:len(x))
articulos_frecuentes
```

# Reglas de asociación.

```python
from mlxtend.frequent_patterns import association_rules
```

```python
association_rules(articulos_frecuentes, metric = 'confidence', 
                 min_threshold = 0.5).sort_values('confidence', ascending = False).reset_index(drop=True)
```

```python
association_rules(articulos_frecuentes, metric = 'lift', 
                 min_threshold = 1.01).sort_values('lift', ascending = False).reset_index(drop=True)
```
